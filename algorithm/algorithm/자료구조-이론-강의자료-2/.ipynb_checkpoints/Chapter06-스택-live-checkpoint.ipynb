## 꼭 알아둬야 할 자료 구조: 스택 (Stack)
* 데이터를 제한적으로 접근할 수 있는 구조
  - 한쪽 끝에서만 자료를 넣거나 뺄 수 있는 구조
* 가장 나중에 쌓은 데이터를 가장 먼저 빼낼 수 있는 데이터 구조
  - 큐: FIFO 정책
  - 스택: LIFO 정책 ex)책을 쌓아올렸을 경우 어떤 책부터 먼저 꺼낼것인가? 당연히 마지막에 쌓은 위에 있는 책을 꺼낼것

### 1. 스택 구조
* 스택은 LIFO(Last In, Fisrt Out) 또는 FILO(First In, Last Out) 데이터 관리 방식을 따름
  - LIFO: 마지막에 넣은 데이터를 가장 먼저 추출하는 데이터 관리 정책
  - FILO: 처음에 넣은 데이터를 가장 마지막에 추출하는 데이터 관리 정책

* 대표적인 스택의 활용
  - 컴퓨터 내부의 프로세스 구조의 함수 동작 방식

* 주요 기능
  - push(): 데이터를 스택에 넣기
  - pop(): 데이터를 스택에서 꺼내기

### 2. 스택 구조와 프로세스 스택
- 스택 구조는 프로세스 실행 구조의 가장 기본
  - 함수 호출시 프로세스 실행 구조를 스택과 비교해서 이해 필요
  
> 엑셀로 이해해보기

```python
# 재귀 함수
def recursive(data):
    if data < 0:
        # 최종적으로 else구문에서 -1이 되어야만 해당 구문이 실행
        print ("ended")
    else:
        #data 변수 한번 출력
        print(data)
        # 출력 후 data 변수를 -1 한다
        recursive(data - 1)
        # 출력 후 -1 한 data를 출력
        print("returned", data)        
```

```python
recursive(4)
```

### 3. 자료 구조 스택의 장단점
- 장점
  - 구조가 단순해서, 구현이 쉽다.
  - 데이터 저장/읽기 속도가 빠르다.
- 단점 (일반적인 스택 구현시) 
  - 데이터 최대 갯수를 미리 정해야 한다. 
    - 파이썬의 경우 재귀 함수는 1000번까지만 호출이 가능함
  - 저장 공간의 낭비가 발생할 수 있음
    - 미리 최대 갯수만큼 저장 공간을 확보해야 함

> 스택은 단순하고 빠른 성능을 위해 사용되므로, 보통 배열 구조를 활용해서 구현하는 것이 일반적임.
> 이 경우, 위에서 열거한 단점이 있을 수 있음

### 4. 파이썬 리스트 기능에서 제공하는 메서드로 스택 사용해보기
* append(push), pop 메서드 제공

```python
# 변수 리스트 초기화
data_stack = list()

# 리스트에 1,2 순서대로 추가
data_stack.append(1)
data_stack.append(2)
```

```python
data_stack
```

```python
# 리스트의 데이터를 삭제 2부터 삭제됨
data_stack.pop()
```

### 5. 프로그래밍 연습 

<div class="alert alert-block alert-warning">
<strong><font color="blue" size="3em">연습1: 리스트 변수로 스택을 다루는 pop, push 기능 구현해보기 (pop, push 함수 사용하지 않고 직접 구현해보기)</font></strong><br>

</div>

```python
# 리스트 초기화
stack_list = list()

# 리스트 푸시하는 함수 추가
def push(data):
    # 인자로 data라는 값을 받아서 리스트 append 시킨다(뒤로 계속 추가 됨)
    stack_list.append(data)

# 리스트에서 데이터를 삭제하는 함수
def pop():
    # data라는 변수에 리스트의 맨끝의 값을 저장
    data = stack_list[-1]
    # 리스트의 마지막 값을 삭제
    del stack_list[-1]
    # data라는 변수에 담은 리스트의 마지막값을 리턴
    return data
```

```python
# 0~9까지 저장
for index in range(10):
    push(index)
```

```python
# 리스트의 마지막값을 출력 후 삭제하기 함수 실행
pop()
```
